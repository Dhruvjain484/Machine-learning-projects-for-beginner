#Importing the Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Importing the dataset


```
dataset = pd.read_csv('winequality-white.csv')
for lables in dataset.columns[:-1]:
  plt.boxplot([dataset[dataset['quality']==i][lables] for i in range(1,11)])
  plt.title(lables)
  plt.xlabel('quality')
  plt.ylabel(lables)
  plt.show()
```

#Data Preprocessing

```
dataset.head()
```

```
dataset.describe()
```

```
plt.figure(figsize = (12, 7))
sns.heatmap(dataset.corr(), annot = True)
```

```
dataset.hist(figsize = (15, 15), bins = 10);
```

```
bins = [0, 5.5, 7.5, 10]
labels = [0,1,2]
dataset['quality'] = pd.cut(dataset['quality'], bins = bins , labels=labels)
dataset.head()
```

```
# dataset = dataset.drop_duplicates()
# dataset = dataset.drop(['free sulfur dioxide'], axis = 1)
# dataset = dataset.drop('density', axis = 1)
```

```
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

#Train test split

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train ,y_test = train_test_split(x, y , test_size = 0.2, random_state = 0)
```

#Training the classification model

```
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(x_train, y_train)
```

#Predicting the result

```
y_pred = classifier.predict(x_test)
y_test = np.array(y_test)
np.concatenate((y_test.reshape(len(y_test), 1), y_pred.reshape(len(y_pred), 1)), 1)
```

```
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

#K Fold Cross validation

```
from sklearn.model_selection import cross_val_score
cvs = cross_val_score(classifier, x_train, y_train)
print(cvs.mean())
print(cvs.std())
```
